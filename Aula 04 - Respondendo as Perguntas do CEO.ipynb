## 0. Loading Data

```python
# Carregando das bibliotecas
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import gridspec
import plotly
import plotly.express as px
pd.set_option('display.float_format', lambda x: '%.2f ' % x)
%matplotlib inline
```

```python
# Lendo o arquivo pelo read_csv
df = pd.read_csv('kc_house_data.csv', low_memory=False, delimiter=',', quotechar='"')
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df.head()
```

## 1. Qual a quantidade de imóveis por nível?

    - Nível 2: Preço entre 450.000 e 645.000
    - Nível 3: Preço acima de 645.000

```python
df['level'] = df['price'].apply(lambda x: 0 if x < 321950 else
                                            1 if (x >= 321950) & (x < 450000) else
                                               2 if (x >= 450000) and (x < 650000) else 3 )
```

## 2. Adicione as seguintes informações no imóvel:
    1. O nome da Rua
    2. O número do Imóvel
    3. O nome do Bairro
    4. O nome da Cidade
    5. O nome do Estado

```python
from geopy.geocoders import Nominatim
```

```python
# Create empty rows
df['road'] = ''
df['house_number'] = ''
df['city'] = ''
df['country'] = ''
df['state'] = ''

# Initialize Nominatim API
geolocator = Nominatim(user_agent='geoapiExercices')

for i in range(len(df)):
    print('Loop: {}/{}'.format(i, len(df)))
    query = str(df.loc[i, 'lat']) + ',' + str(df.loc[i, 'long'])
    
    # API Request
    response = geolocator.reverse(query)
    
    # Populate Data
    if 'house_number' in response.raw['address']:
        df.loc[i, 'house_number'] = response.raw['address']['house_number']
        
    if 'road' in response.raw['address']:
        df.loc[i, 'road'] = response.raw['address']['road']
        
    if 'neighbourhood' in response.raw['address']:
        df.loc[i, 'neighbourhood'] = response.raw['address']['neighbourhood']
        
    if 'city' in response.raw['address']:
        df.loc[i, 'city'] = response.raw['address']['city']
    
    if 'country' in response.raw['address']:
        df.loc[i, 'country'] = response.raw['address']['country']
        
    if 'state' in response.raw['address']:
        df.loc[i, 'state'] = response.raw['address']['state']
```

## 3. Adicione o nível do imóvel no mapa como uma cor

```python
df['level'] = df['level'].astype(str)
houses = df[['id', 'lat', 'long', 'price', 'level']].copy()

fig = px.scatter_mapbox(houses,
                       lat = 'lat',
                       lon = 'long',
                       color = 'level',
                       size = 'price',
                       color_continuous_scale = px.colors.cyclical.IceFire,
                       size_max=15,
                       zoom = 10)

fig.update_layout(mapbox_style='open-street-map')
fig.update_layout(height=600, margin={'r':0, 't':0, 'l':0, 'b':0})
fig.show()
```

## 5. Adicione opções de filtros para eu fazer minhas próprias análises:
    1. Eu quero escolher visualizar imóveis com vista para água ou não;
    2. Eu quero escolher filtrar os imóveis até um certo valor de preço

```python
import ipywidgets as widgets
from ipywidgets import fixed
```

```python
df['iswaterfront'] = df['waterfront'].apply(lambda x: 'yes' if x == 1 else 'no')
df['level'] = df['price'].apply(lambda x: 0 if x < 321950 else
                                            1 if (x >= 321950) & (x < 450000) else
                                               2 if (x >= 450000) and (x < 650000) else 3 )
df['level'] = df['level'].astype(int)
style = {'description':'initial'}

# Interative buttons
price_limit = widgets.IntSlider(
    value = 540000,
    min = 75000,
    max = 7700000,
    step = 1,
    description = 'Maximum Price',
    disable = False,
    style = style)
```

```python
waterfront_bar = widgets.Dropdown(
    options=['yes', 'no'],
    values='yes',
    description='Water View',
    disable=False)


def update_map(df, waterfront, limit):
    houses = df[(df['price'] <= limit) & 
                            (df['iswaterfront'] == waterfront)][['id', 'lat', 'long', 'price', 'level']]
    fig = px.scatter_mapbox(houses,
                           lat = 'lat',
                           lon = 'long',
                           color = 'level',
                           size = 'price',
                           color_continuous_scale = px.colors.cyclical.IceFire,
                           size_max=15,
                           zoom = 10)

    fig.update_layout(mapbox_style='open-street-map')
    fig.update_layout(height=600, margin={'r':0, 't':0, 'l':0, 'b':0})
    fig.show()
```

```python
widgets.interactive(update_map, df=fixed(df), waterfront=waterfront_bar, limit=price_limit)
```

## 6. Adicionar opções de filtro no último dashboard enviado:
    1. Eu quero escolher visualizar valores a partir de uma data disponível para comprar

```python
# Change date format
df = pd.read_csv('kc_house_data.csv', low_memory=False, delimiter=',', quotechar='"')
df['year'] = pd.to_datetime(df['date']).dt.strftime('%Y')
df['date'] = pd.to_datetime(df['date']).dt.strftime('%Y-%m-%d')
df['year_week'] = pd.to_datetime(df['date']).dt.strftime('%Y-%U')

# Widgets control date
date_limit = widgets.SelectionSlider(
    options=df['date'].sort_values().unique().tolist(),
    value = '2014-12-01',
    description = 'Disponível',
    continuous_update = False,
    orientation='horizontal',
    readout=True)
```

```python
def update_map(df, limit):
#     Filter Data
    df = df[df['date'] >= limit].copy()
    
    fig = plt.figure(figsize=(21,12))
    specs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)
    
    ax1 = fig.add_subplot(specs[0, :])
    ax2 = fig.add_subplot(specs[1, 0])
    ax3 = fig.add_subplot(specs[1, 1])
    
    by_year = df[['id', 'year']].groupby('year').sum().reset_index()
    ax1.bar(by_year['year'], by_year['id'])
    
    by_day = df[['id', 'date']].groupby('date').mean().reset_index()
    ax2.plot(by_day['date'], by_day['id'])
    ax2.set_title('Title: Avg Price by Day')
    
    by_week = df[['id', 'year_week']].groupby('year_week').mean().reset_index()
    ax3.bar(by_week['year_week'], by_week['id'])
    ax3.set_title('Title: Avg Price by Week of the Year')
    
    plt.xticks(rotation=60);
```

```python
widgets.interactive(update_map, df=fixed(df), limit=date_limit)
```
